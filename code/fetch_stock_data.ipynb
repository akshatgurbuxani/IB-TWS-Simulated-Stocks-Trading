# Import Libraries

```python
import yfinance as yf
import pandas as pd
```

# Fetch stock data from Yahoo Finance

```python
ticker = 'ES=F'   # Represents E-mini S&P
emini = yf.Ticker(ticker)
   
start_date = '2022-01-01'
end_date = '2024-03-30'

stock_data = emini.history(start=start_date, end = end_date, interval='1d', back_adjust=True, auto_adjust=True, prepost=True)
```

```python
stock_data
```

```python
stock_data.index = pd.to_datetime(stock_data.index.date)
stock_data.index.name = "Date"
```

```python
stock_data
```

```python
base_dir = "../Data"
stock_data.to_csv(f"{base_dir}/stock_data_.csv")
```

# Download data from Yahoo Finance

```python
stock_data = yf.download(ticker, period="5y")

base_dir = "../Data"
stock_data.to_csv(f"{base_dir}/stock_data.csv")
```
